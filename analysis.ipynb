```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data import

```python
df = pd.read_csv('Data_Benzene_Reduced.csv', sep=";", header=None)
```

```python
df.rename(columns={0: 'benzene', 1: 'eq_ratio'}, inplace=True)
```

# Task 1

## Benzene

```python
df_by_benzene = df.groupby('benzene', as_index=False).mean()
```

```python
df_by_benzene
```

```python
grid_shape = (6, 8)
```

### Heatmaps

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(30, 30), sharey=True)
fig.set_facecolor('white')
# find min and max values to get the same scale
min_val = df_by_benzene.iloc[:,2:].min().min()
max_val = df_by_benzene.iloc[:,2:].max().max()
for index, value in enumerate(df_by_benzene["benzene"]):
    T_data = np.flip(df_by_benzene[df_by_benzene["benzene"] == value].iloc[:,2:].to_numpy().reshape(grid_shape), axis=0)
    sns.heatmap(T_data, cmap="YlOrBr", ax=ax.flatten()[index], vmin=min_val, vmax=max_val, annot=True, fmt=".2f")
    ax.flatten()[index].set_title(f"Grid temperatures for {value} benzene percentage")
```

### Histograms

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(20, 20), sharey=True)
fig.set_facecolor('white')

for index, benzene_p in enumerate(df["benzene"].unique()):
    # array of temperatures for each benzene percentage
    T_data = np.flip(df[df["benzene"] == benzene_p].iloc[:,2:].to_numpy().flatten(), axis=0)
    # plot the histogram
    sns.histplot(T_data, ax=ax.flatten()[index], kde=False, bins=50)
    ax.flatten()[index].set_title(f"Temperature distribution for {benzene_p} benzene percentage")
```

### Box plots

```python
fig = plt.figure(figsize=(12,8))
# put all temperatures in a single column
benzene_boxplot_df = df.melt(id_vars=("benzene", "eq_ratio"), var_name="grid_position", value_name="temperature")
sns.boxplot(x="benzene", y="temperature", data=benzene_boxplot_df)
plt.xlabel("Benzene percentage")
plt.ylabel("Temperature")
plt.title("Boxplot of temperature for benzene percentage")
```
